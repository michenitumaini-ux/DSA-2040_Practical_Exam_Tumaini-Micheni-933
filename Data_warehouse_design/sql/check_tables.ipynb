```python
import sqlite3

# Connect to the database
conn = sqlite3.connect('retail_dw.db')
cursor = conn.cursor()

# List all tables
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print("Tables in the database:", cursor.fetchall())

# Show schema for SalesFact
cursor.execute("PRAGMA table_info(SalesFact);")
print("\nSalesFact table schema:")
for col in cursor.fetchall():
    print(col)

# Show schema for CustomerDim
cursor.execute("PRAGMA table_info(CustomerDim);")
print("\nCustomerDim table schema:")
for col in cursor.fetchall():
    print(col)

# Show schema for ProductDim
cursor.execute("PRAGMA table_info(ProductDim);")
print("\nProductDim table schema:")
for col in cursor.fetchall():
    print(col)

# Show schema for TimeDim
cursor.execute("PRAGMA table_info(TimeDim);")
print("\nTimeDim table schema:")
for col in cursor.fetchall():
    print(col)

conn.close()
```
